```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
coin_tosses = np.random.binomial(n=5,p=0.5,size=10)
print(coin_tosses)
frequency = np.count_nonzero(coin_tosses)
print(frequency)
relative_frequency = frequency/coin_tosses.shape[0]
print(relative_frequency)
```

```python
plt.xlabel("NUmber of desired outcomes out of 5 BInomial Trials")
plt.ylabel("P(k/5)")
plt.title(f"Binomial Distribution")
plt.bar([0,1,2,3,4,5],[0.03125,0.15625,0.3125,0.3125,0.15625,0.03125])
```
